```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
import numpy as np
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])


print("=== zomato sales analysis ===")
print("Shape of the data:",sales_data.shape)
print("\n Sample data for first 3 restaurant:",sales_data[:3])
print("Yeh data using range and all",sales_data[:,1:])
```

```python
#total sales per year
import matplotlib.pyplot as plt
print(np.sum(sales_data[:,1:],axis=0))#ye vertical addition h
print(np.sum(sales_data[:,1:],axis=1))#this was horizontal addition 

#minimum sale by a company in the year
print("This is the minimum sale ",np.min(sales_data[:,1:],axis=1))
print("Maximum sales per year is ", np.max(sales_data[:,1:],axis=0))

#average sales per year
print("Average sales per year:",np.average(sales_data[:,1:],axis=0))
print("Average sales per restaurant:",np.average(sales_data[:,1:],axis=1))

#cumulative sales aka cumsum
print("The cumulative sales of the restaurant:",np.cumsum(sales_data[:,1:],axis=1))
cumsum=np.cumsum(sales_data[:,1:],axis=1)






## Matplotlib 

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales per year Across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
